```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
X = 2 * np.random.rand(n, 1)
y = 4 + 3 * X + np.random.randn(n, 1)
```

```python
plt.plot(X, y, "b.")
```

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
X_new = np.linspace(0, 2, n).reshape(n,1)
y_p = lr.predict(X_new)
```

```python
plt.plot(X,y,"b.", label="Data points")
plt.plot(X_new,y_p,"r-", linewidth=2, label="Predictions")
plt.axis([0,2,2,12])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("y", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```

### RidgeRegression

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=1, solver='cholesky')
```

```python
rr.fit(X,y)
```

```python
rr.predict([[1.5]])
```

```python
X_r = np.linspace(0, 2, n).reshape(n,1)
y_r = rr.predict(X_new)
```

```python
plt.plot(X,y,"b.")
plt.plot(X_new,y_p,"r-", linewidth=2, label="Least Square")
plt.plot(X_r,y_r,"g-", linewidth=2, label="Ridge")
plt.axis([0,2,2,12])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("y", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
la_r = Lasso(alpha=0.1)
```

```python
la_r.fit(X,y)
```

```python
X_la = np.linspace(0, 2, n).reshape(n,1)
y_la = la_r.predict(X_new)
```

```python
plt.plot(X,y,"b.")
plt.plot(X_new,y_p,"r--", linewidth=2, label="Least Square")
plt.plot(X_r,y_r,"g-", linewidth=2, label="Ridge")
plt.plot(X_la,y_la,"c-", linewidth=1, label="Lasso")
plt.axis([0,2,2,12])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("y", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```
