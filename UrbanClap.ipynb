**Question  4 :**<br> Get Month wise New & Old Pro count<br>
New Pro = First Job in the month (only consider 2020)<br>
Old Pro  = First Job in some month in past (only 2020)<br>

```python
data_BDATE_PROVIDER_ID = data[['BDATE', 'PROVIDER_ID']].sort_values(['BDATE','PROVIDER_ID']).drop_duplicates()
```

```python
PROVIDER_ID_PRO = [list(data_BDATE_PROVIDER_ID[data_BDATE_PROVIDER_ID['BDATE'] == i]['PROVIDER_ID']) for i in data_BDATE_PROVIDER_ID.BDATE.unique()]
```

```python
COLL, NEW_PRO, REPEAT_PRO  = [], [], []
for i in PROVIDER_ID_PRO:
    new_addr = []
    old_addr = []
    for j in i:
        if j not in COLL:
            new_addr.append(j)
        else:
            old_addr.append(j)     
    NEW_PRO.append(new_addr)
    REPEAT_PRO.append(old_addr)
    COLL.extend(NEW_PRO[0])
```

```python
newdata_PRO_OLD = pd.DataFrame()
newdata_PRO_OLD['MTH'] = newdata['MTH']
newdata_PRO_OLD['NEW_PRO'] = [len(i) for i in NEW_PRO]
newdata_PRO_OLD['REPEAT_PRO'] = [len(i) for i in REPEAT_PRO]
```

New Pro = First Job in the month (only consider 2020)<br>
Old Pro  = First Job in some month in past (only 2020)<br>


```python
dates = data['BDATE'].unique()
```

```python
newdata_PRO_OLD['NEW_PRO_DELIVERED'] = [sum(data[data['BDATE'] == i][data[data['BDATE'] == i]['PROVIDER_ID'].isin(j)][['CUSTOMER_REQUEST_ID','PROVIDER_ID']].groupby('PROVIDER_ID').count()['CUSTOMER_REQUEST_ID']) for i, j in zip(dates, NEW_PRO)]
newdata_PRO_OLD['OLD_PRO_DELIVERED'] = [sum(data[data['BDATE'] == i][data[data['BDATE'] == i]['PROVIDER_ID'].isin(j)][['CUSTOMER_REQUEST_ID','PROVIDER_ID']].groupby('PROVIDER_ID').count()['CUSTOMER_REQUEST_ID']) for i,j in zip(dates, REPEAT_PRO)]
```

```python
newdata_PRO_OLD['NEW_PRO_UTIL'] = newdata_PRO_OLD['NEW_PRO_DELIVERED'] / newdata_PRO_OLD['NEW_PRO']
newdata_PRO_OLD['OLD_PRO_UTIL'] = newdata_PRO_OLD['OLD_PRO_DELIVERED'] / newdata_PRO_OLD['REPEAT_PRO']
```
